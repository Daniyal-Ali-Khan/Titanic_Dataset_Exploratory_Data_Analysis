```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df=pd.read_csv('titanic_dataset.csv')
```

```python
# Display the first few rows of the dataset
df.head()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Summary statistics
df.describe()
```

```python
# Distribution of numerical features
df.hist(bins=20, figsize=(15, 10))
plt.show()
```

```python
# Correlation heatmap
correlation_matrix = df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.show()
```

```python
# Countplot for categorical variables
plt.figure(figsize=(12, 6))
sns.countplot(x='2urvived', data=df)
plt.title('Survival Count (0: Did not survive, 1: Survived)')
plt.show()
```

```python
# Boxplot to visualize age distribution by class
plt.figure(figsize=(12, 6))
sns.boxplot(x='Pclass', y='Age', data=df)
plt.title('Age distribution by Pclass')
plt.show()
```

```python
# Scatter plot for Fare vs Age
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Age', y='Fare', data=df)
plt.title('Scatter plot of Age vs Fare')
plt.show()
```

```python
# Pairplot for selected variables
sns.pairplot(df[['Age', 'Fare', 'Pclass', '2urvived']], hue='2urvived')
plt.show()
```

```python
# Bar plot for the distribution of passengers by gender
plt.figure(figsize=(10, 5))
sns.countplot(x='Sex', data=df)
plt.title('Distribution of Passengers by Gender')
plt.show()
```

```python
# Bar plot for the distribution of passengers by class
plt.figure(figsize=(10, 5))
sns.countplot(x='Pclass', data=df)
plt.title('Distribution of Passengers by Class')
plt.show()
```
